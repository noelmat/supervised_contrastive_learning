```python
def get_splits(FOLD, x):
    return splits[FOLD]

def get_data(labels, bs=64, presize=500, resize=384):
    tfms = [Rotate(90), Warp(magnitude=0.3),Zoom(min_zoom=0.9, max_zoom=1.3), Brightness(max_lighting=0.6), Flip(), Contrast(0.4), Resize(resize)]
    comp = setup_aug_tfms(tfms)
    return DataBlock(blocks=(ImageBlock, CategoryBlock),
        get_items=lambda p: get_image_files(p),
        get_y=partial(get_label, labels),
        splitter=partial(get_splits, 0),
        item_tfms=[Resize(presize)],
        batch_tfms=[*comp, Normalize.from_stats(*imagenet_stats)]).dataloaders(path/'train_images',bs=bs,num_workers=8, shuffle_train=False)
```

```python
def normalize(x):
    return F.normalize(x, p=2, dim=1)
```

```python
class Model(Module):
    def __init__(self, pretrained=True):
        self.base_model = nn.Sequential(*list(models.resnet50(pretrained=pretrained).children())[:-1])
        self.proj = nn.Sequential(
            nn.Linear(2048, 2048), 
            nn.ReLU(),
            nn.Linear(2048, 128)
        )
    
    def forward(self, x1, x2):
        out1 = self.base_model(x1)
        out2 = self.base_model(x2)
        out1 = normalize(out1).squeeze()
        out2 = normalize(out2).squeeze()
        return normalize(self.proj(out1)), normalize(self.proj(out2))
```

```python
def supconloss(y, o1, o2):
    tot_loss = 0
    for i in range(y.shape[0]):
        p = torch.where(y == y[i])[0]
        p = torch.where(p != i)[0]
        P_i = len(p)

        a = torch.where(y != y[i])[0]
        numerator = ((o1[i][None, :] @ o2[p].T)/tau).exp()
        deno = ((o1[i][None, :] @ o2[a].T)/tau).exp().sum()
        loss = (numerator/deno).log().sum()
        loss = loss * -1/(P_i+1)
        tot_loss += loss    
    return tot_loss
```

```python
class SupConLearner(Learner):
    def all_batches(self):
        self.n_iter = len(self.dl)
        self.dl1 = self.dl
        self.dl2 = self.dl
        for o in enumerate(zip(self.dl1, self.dl2)):
            self.one_batch(*o)
            
    def _split(self, b):
        i = getattr(self.dls, 'n_inp', 1 if len(b)==1 else len(b)-1)
        b1 = b[0]
        b2 = b[1]
        
        xb, yb = b1[:i][0],b1[i:][0]
        self.xb = torch.cat((xb,b2[:i][0]))
        self.yb = torch.cat((yb,b2[i:][0]))
```
