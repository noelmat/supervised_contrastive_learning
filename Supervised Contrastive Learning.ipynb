```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(n_splits=5)
splits = {}

for i,(tr_idx,val_idx) in enumerate(cv.split(train_df, train_df['label'])):
    splits[i] = [tr_idx, val_idx]
```

```python
df = train_df.set_index('image_id')
labels = df.to_dict()['label']
```

```python
def get_label(labels, x):
    x = Path(x)
    return labels[x.name]

fn = train_df.iloc[0,0]
```

```python
%%time
get_label(labels,path/'train_images'/fn)
```

```python
def get_splits(FOLD, x):
    return splits[FOLD]

def get_data(labels, bs=64, presize=500, resize=384):
    tfms = [Rotate(90), Warp(magnitude=0.3),Zoom(min_zoom=0.9, max_zoom=1.3), Brightness(max_lighting=0.5), Flip(), Contrast(0.4), Resize(resize)]
    comp = setup_aug_tfms(tfms)
    return DataBlock(blocks=(ImageBlock, CategoryBlock),
        get_items=lambda p: get_image_files(p),
        get_y=partial(get_label, labels),
        splitter=partial(get_splits, 0),
        item_tfms=[Resize(presize)],
        batch_tfms=[*comp, Normalize.from_stats(*imagenet_stats)]).dataloaders(path/'train_images',bs=bs,num_workers=8)
```

```python
dls = get_data(labels, bs=16)
```

```python
dls.show_batch()
```

```python
x, y = dls.one_batch()
```

```python
def normalize(x):
    return (x - x.mean())/(x.min() - x.max())
```

```python
class Model(Module):
    def __init__(self, pretrained=True):
        self.base_model = nn.Sequential(*list(models.resnet50(pretrained=pretrained).children())[:-1])
        self.proj = nn.Sequential(
            nn.Linear(2048, 2048), 
            nn.ReLU(),
            nn.Linear(2048, 128)
        )
    
    def forward(self, x1, x2):
        out1 = self.base_model(x1)
        out2 = self.base_model(x2)
        out1 = normalize(out1).squeeze()
        out2 = normalize(out2).squeeze()
        return normalize(self.proj(out1)), normalize(self.proj(out2))
```

```python
model = Model()
```

```python
o1, o2 = model(x, x)
```

```python
tau = 0.1
```

```python
def supconloss(y, o1, o2):
    tot_loss = 0
    for i in range(y.shape[0]):
        p = torch.where(y == y[i])[0]
        p = torch.where(p != i)[0]
        P_i = len(p)

        a = torch.where(y != y[i])[0]
        numerator = ((o1[i][None, :] @ o2[p].T)/tau).exp()
        deno = ((o1[i][None, :] @ o2[a].T)/tau).exp().sum()
        loss = (numerator/deno).log().sum()
        loss = loss * -1/(P_i+1)
        tot_loss += loss    
    return tot_loss
```

```python
%%time
supconloss(y, o1, o2)
```
